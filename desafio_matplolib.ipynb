```python
import pandas as pd
import matplotlib.pyplot as plt
dados = pd.read_csv('./athlete_events.csv')
dados.head()
```

Criar um dataset para filtrar apenas as colunas com peso e altura do sexo masculino

```python
masculinos = dados.loc[dados['Sex']=='M']
masculinos.head()
```

Criando a relação altura e peso

```python
altura = masculinos['Height']
peso = masculinos['Weight']
```

Criando o gráfico

```python
plt.scatter(altura,peso)
plt.show()
```
